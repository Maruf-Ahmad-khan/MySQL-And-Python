```python
def Read_data(dataset):

  try:

    ans = spark.read.csv(dataset, header = True, inferSchema = True)
    return ans

  except TypeError:
          raise TypeError('Type Error is occurred:')

  except Exception as e:
          print(f"Error: {e}")

ans = Read_data(dataset = "/content/Practice_data.csv")
ans.show()
```

```python
# 1. Find records where 'Amount'
# is greater than the average 'Amount' for the entire dataset.
import pandas as pd
df1 = pd.read_csv(r'/content/Practice_data.csv')
df1.head()
```

```python
df1[df1['Amount'] > df1['Amount'].mean()]
```

```python
# 2. Retrieve entries where the 'Payout' is equal to the maximum 'Payout'.
df1[df1['Payout'] == df1['Payout'].max()]
```

```python
# 3. Select rows where the 'Date' is the earliest in the dataset.
df1[df1['Date'] == df1['Date'].min()]
```

```python
from pyspark.sql.functions import col, avg
```

```python
filtered_df = ans.filter(col("Amount") > ans.agg(avg("Amount")).collect()[0][0])
filtered_df.show()
```

```python
# 4. List campaigns where the 'Amount'
#  is less than the average 'Amount' of all campaigns.

df1[df1['Amount'] < df1['Amount'].mean()]
```

```python
# 5. Identify records where 'Payout' equals
#  the minimum 'Payout' for campaigns with "pending" status.

df1[df1['Payout'] == df1[df1['Status'] == 'pending']['Payout'].min()]
```

```python

# 6. Fetch details of
# campaigns with the highest total 'Amount'.
df1.loc[df1['Amount'] == df1['Amount'].max()]
```

```python
# 7. Find entries where 'Payout' matches the median 'Payout' for the dataset.
df1[df1['Payout'] == df1['Payout'].median()]
```

```python
# 8. Select rows where the 'Sub_Id' occurs most frequently.
df1[df1['Sub_Id'] == df1['Sub_Id'].mode()[0]]
```

```python
# 9. Retrieve records where the
# 'Date' equals the most recent date for campaigns named "Ajio".

df1[(df1['Campaign_Name'] == 'Ajio') & (df1['Date'] == df1[df1['Campaign_Name']=='Ajio']['Date'].max())]
```

```python
#  Find records with the smallest 'Id' for each 'Status'.
df1.groupby('Status').agg(Smallest_Id = ('Id', 'min'))
```
